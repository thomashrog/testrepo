As requested in the homework, the code below: uses bootstrap code to illustrate the central limit theorem in R markdown and push the result to GitHub. 
- Part A: Use a normal distribution with two different sample sizes and; 
- Part B: An exponential distribution with two different sample sizes. 

### Part A: Normal Distribution with Two Different Sample Sizes

```r
#Creating a dataset with 200 observations, a mean of 100 and a standard deviation of 10

rnormsample1 <- round(rnorm(200, 100, 10))
rnormsample2 <- round(rnorm(1200, 20, 20))

#Creating 30 bootstrap samples for sample #1
bssamples1 <- lapply(1:30, function(i)
sample(rnormsample1, replace = T))
bssamples1[1]

#Creating 30 bootstrap samples for sample #2
bssamples2 <- lapply(1:30, function(i)
sample(rnormsample2, replace = T))
bssamples2[1]
    
```

### Calculate the Median Values of the Bootstrap Distribution for Plotting

```r
# Find the median value of sample #1
r.bootmed1 <- sapply(bssamples1, median)
r.bootmed1

# Find the median value of sample #2
r.bootmed2 <- sapply(bssamples2, median)
r.bootmed2
```

### Plotting the Distribution

```r
hist(r.bootmed1)
hist(r.bootmed2)
```

### Part B: Exponential Distribution with Two Different Sample Sizes

```r
# Creating the first exponential distribution

expdist1 =  rexp(n = 4000, rate = .5)
matrixexpdist1 = matrix(expdist1, nrow=100,ncol=5)
means.exp1 = apply(matrixexpdist1,1,mean)

# Creating the second exponential distribution
expdist2 =  rexp(n = 2000, rate = .1)
matrixexpdist2 = matrix(expdist2, nrow=800,ncol=10)
means.exp2 = apply(matrixexpdist2,1,mean)
```

### Plotting the Distribution

```r
hist(means.exp1)
hist(means.exp2)
```
